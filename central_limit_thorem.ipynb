# Central Limit Theorem
The central limit theorem states that if we take large number of samples from any population with finite mean and variance then the distribution of the sample means will follow the normal distribution regardless of the type of the original distribution. 

Also the mean of these sample means will be equal to the population mean and standard error(standard deviation of the sample means) will decrease with increase in sample size.

The **central limit theorem** states that as you take more and more random samples from a population, the means you calculate from the samples will be approximately normally distributed, regardless of the shape of the original population distribution. This normal distribution of the sample means is an indirect way of claiming normality for any dataset with a sufficiently large number of observations typically $n>30$.

```python
import numpy as np
import matplotlib.pyplot as plt

# Number of samples to draw
num_samples = [1, 10, 50, 100]

# List to store sample means
means = []

# Setting a random seed for reproducibility
np.random.seed(1)

# Generating random numbers, computing their means, and storing in means list
for j in num_samples:
    sample_means = [
        np.mean(np.random.randint(-40, 40, j)) for _ in range(1000)
    ]
    means.append(sample_means)

# Plotting all the means in one figure
fig, ax = plt.subplots(2, 2, figsize=(8, 8), sharey='row')

for k in range(4):
    i, j = divmod(k, 2)
    ax[i, j].hist(means[k],
                  bins=25,
                  density=True,
                  alpha=0.75,
                  color='darkgreen',
                  edgecolor='black')
    ax[i, j].set_title(f'Sample size: {num_samples[k]}')
    
    # Set x-labels only for the bottom subplots
    if i == 1:
        ax[i, j].set_xlabel('Sample Mean')

    # Set y-labels only for the left subplots
    if j == 0:
        ax[i, j].set_ylabel('Density')

    # Remove labels for the inner plots
    for i in range(2):
        for j in range(2):
            if not (i == 1 or j == 0):
                ax[i, j].set_xlabel('')
                ax[i, j].set_ylabel('')

plt.tight_layout()
plt.show()
```
